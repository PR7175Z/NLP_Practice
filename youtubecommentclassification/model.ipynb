```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import nltk
```

```python
nltk.download('all')
```

```python
dataframe = pd.read_csv('data/YoutubeCommentsDataSet.csv')
```

```python
dataframe
```

```python
from nltk import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
stopwords = set(stopwords.words('english'))
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
X = dataframe['Comment'].astype(str)
```

```python
X
```

```python
def preprocess_text(cmt):
    if not isinstance(cmt, str):
        return 
    
    tokenize = word_tokenize(cmt)
    stemmed = [ps.stem(x) for x in tokenize if x not in stopwords ]

    return  ' '.join(stemmed)
```

```python
processed_text = {}
```

```python
processed_text['comments'] = X.apply(preprocess_text)
processed_text['Sentiment'] = dataframe['Sentiment']
```

```python
display(processed_text)
```

```python
! pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.naive_bayes import MultinomialNB
```

```python
vectorizer = TfidfVectorizer(stop_words='english')

X_vect = vectorizer.fit_transform(X)

X_vect
```

```python
sentiment_mapping = {'positive': 1, 'negative': 0, 'neutral': 2}
y = processed_text['Sentiment'].map(sentiment_mapping)
```

```python
print(y)
```

```python
xtrain,xtest, ytrain, ytest = train_test_split(X_vect, y, train_size=0.8, test_size=0.2, random_state=42)
```

```python
model = MultinomialNB()
model.fit(xtrain, ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
print(accuracy_score(ytest, ypred))
```

```python
print(confusion_matrix(ytest, ypred))
```

```python
cm = confusion_matrix(ytest, ypred)

# Plot the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()
```

```python
print(classification_report(ytest, ypred))
```
