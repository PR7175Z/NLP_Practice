<h2>Constituency parsing</h2>
process of analyzing a sentence and identifying its constituent parts (such as noun phrases, verb phrases, etc.) and their hierarchical structure

```python
import benepar
from nltk import Tree
import spacy
```

```python
benepar.download('benepar_en3')
```

```python
constituencyparsing = spacy.load("en_core_web_sm")
constituencyparsing.add_pipe("benepar", config={"model": "benepar_en3"})

doc = constituencyparsing("The quick brown fox jumps over the lazy dog.")
for sent in doc.sents:
    print(sent._.parse_string)
```

```python
doc = constituencyparsing("This world is full of people who don't want to work.")
sent = list(doc.sents)[0]
```

```python
tree = Tree.fromstring(sent._.parse_string)
```

```python
tree.pretty_print()
```

<h2>Dependency Parsing</h2>

breaks down a sentence into a set of words linked by syntactic relationships, showing how each word depends on another.
<h3>Key Concepts</h3>
<ul>
    <li><strong>Token</strong>: Each word in sentence (including punctuations)</li>
    <li><strong>Head</strong>: The word that a token depends on.</li>
    <li><strong>Dependency Relation</strong>: The grammatical relationship between the token and its head.</li>
</ul>

```python
from spacy import displacy
```

```python
dependencyparsing = spacy.load("en_core_web_sm")
doc = dependencyparsing("This world is full of people who don't want to work.")

for token in doc:
    print(f"{token.text} {token.dep_} --> {token.head.text}")
```

```python
def to_nltk_tree(node):
    if node.n_lefts + node.n_rights > 0:
        return Tree(node.orth_, [to_nltk_tree(child) for child in node.children])
    else:
        return node.orth_

root = [token for token in doc if token.head == token][0]
tree = to_nltk_tree(root)
tree.pretty_print()
```
